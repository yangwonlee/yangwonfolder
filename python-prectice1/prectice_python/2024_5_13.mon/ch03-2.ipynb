```python
# while문 학습(반복문(iteration)1)
# 프로그램의 꽃. 반복문은 컴퓨터의 특성이나 진배 없는 것. 이것을 하기 위해 배운다
#라고 할 정도로 중요한 것. 컴퓨터는 인간과는 다르게 대량반복이 가능하기에.
# if 조건문:
#    수행할 문장 1
#    수행할 문장 2

# while 조건문:
#   수행할 문장 1
#   수행할 문장 2
#   수행할 문장 3

#ㄴ 이프문은 코드 생성시 문장 2 밑으로 내려가지만
# 와일문은 조건문 옆으로 감. 조건문이 true가 되면 계속 반복. 수행할문장123조건문4
# 567 조건문8 로. 이것이 가장 큰 조건문.
```

```python
treeHit = 0     # 트리힛이 0이면, (마지막줄) 넘어갑니다.
while treeHit < 10: # 트리힛 10 이하일때
    treeHit = treeHit + 1 #트리힛 값을 1씩 증가한다.
    print("나무를 %d번 찍었습니다." % treeHit) # 나무를x번 찍는다.
    if treeHit == 10:      # 만약 트리힛을 10회가 된다면
        print("넘어 갑니다.")   # 넘어간다.
# 와일문 조건문은 트리히트가 10보다 작은 동안 와일문에 포함된 문장을 계속 수행.
# 트리힛이 10 이상이 되면 코딩 첫줄의 트리힛<10 조건문이 거짓이 되므로 그제서야
#와일문을 빠져나가게 됨.
# 다시금 얘기해서 와일문은 조건문이 들어가고, 반복된다는 것이 중요.
```

```python
# 와일문 만들기 예제.
prompt = """ArithmeticError
    1. Add
    2. Del
    3. List
    4. Quit     

    Enter number:
    """
# 4번이 나올때까지 문장이 종료되지 않음. ""안에, 위에 입력을 하는것에 4번을 누를 때 까지.

number = 0
while number != 4:
    print(prompt)
    number = int(input())       # int= 문자를 정수형으로 변환시키는 함수. 함수 내엔 무조건 숫자만 들어가야 하니.5트

```

```python
# while 강제로 빠져나가기.
# 1. 조건식이 false 일때.
# 2. break를 활용하여 while문 빠져나가기. 이것을 이번 장을 통해서 배울 예정.
# 자판기 안에 커피가 충분히 있을 경우 동전을 넣으면 커피가 나옴. 그러나 자판기가 제대로 작동하려면 커피가 얼마나 남았나
# 항시 검사해야함. 만일 커피가 떨어졌다면 판매를 중단하고 '판매중지' 문구를 사용자에게 보여줘야 함. 이렇게 판매를 강제로
# 멈추게 하는것이 바로 break문.
coffee = 10 # 커피는 양이 10. 10개 있다는 뜻.
money = 300 # 돈은 300원이 있다는 뜻.
while money:    #돈이 300이면 True. 0은 false. 0이외의 숫자는 True. false가 될때까지 와일문을 돌림. 첫라인부터 끝까지
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee - 1
    print("남은 커피의 양은 %d개 입니다." % coffee) # 돈을 받았으니 커피를 줍니다. 남은~.를 이프문이 나올때까지 돌림.

    if coffee == 0:     # 만약 커피의 양이 0이라면,( 반드시 엔터 두번을 눌러야 하는것 잊지말기.)
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")     #커피가 다 떨~를 출력하시오.
         # 코드구현 ㄱ. 브레이크문 이라고 함. 코드 구현이라고 표현함.   모든 제어문은 중첩을 허용. 와일문 상동. 
    break
```

```python
# while 문의 맨 처음으로 돌아가기. break 문과 정 반대되는 continue문.
# 윗 코드를 브레이크에서 컨티뉴로 바꾼다면? 브레이크문은 True를 보지 않고 18번 라인으로 나오지만, 
# 컨티뉴 문은 2번으로 돌아감. 위로 올라감.
a = 0 # a는 0으로 세팅
while a < 10: # a가 10보다 작으면 True. 방금 또 틀렸는데 콜론을 빼먹음.
    a = a + 1
    if a% 2 == 0: continue # a를 2로 나눴을때 나머지가 0이라는 뜻
    print(a) # 6번을 만나면 3번으로 돌아감. 10 이하 1,3,5,7,9. print를 내어쓰기를 한다면? 10이 나올 것. 
# 10이 충족이 되어야 와일문에서 나갈 수 있기 떄문
# 위 코드는 0부터 10까지 1씩 더해서 짝수일 땐 위로 올라가서 다시계산 하고, 홀수만 내보내는 것.
```

```python
#문 7. 손도 못댔음. 답.
a = 1
b = 1
while a < 10:
    print(a,"단")
    while b < 10:   # a= 2 b = 10. 이걸 b를 1로 바꿔주는 작업이 필요함.
        print("{}*{}*{}".format(a,b,a * b))
        b += 1 # 이상태면 a는 증가할때 b는 예전 값을 유지하고 있음. 
    a += 1          # 한차례 도달하면 a = 2, b = 10
    b = 1       # b를 다시 1로 고정시키는 라인
```

```python
# 문제 8
# 피보나치 수열문제
# 1, 1, 2, 3, 5, 8, 13, 21, 45...와 같이 1번항과 2번항은 각각 1로 세팅이 되고 
# 3번 항부터는 바로 앞 두 항의 합으로 계산되는 수열이다. 피보나치 수열의 첫 50개 항을 차례대로 출력하는 프로그램 ㄱ
prev = 0
current = 1
i = 1 # 인덱스 = 1 # 이건 인덱스 카운트.

while i < 51:
    print(current)  # 이건 실제 출력되는 값.
    temp = prev
    prev = current
    current = current + temp        #핵심은 결국 스왑. 나중에 자바 배울때 또 ㄱ

    i += 1
```
