```python
import numpy as np  #  넘파이를 사용할 때 np라는 이름으로 많이 사용(관례)

# numpy로 2차원 배열 만들기.
a = np.array([[2,3],[5,2]]) #객체는 넘파이.
print(a)
print(type(a))

# 이중화 리스트 만들기
a = [[2, 3], [5, 2]]
print(a)
print(type(a))
```

```python
# 넘파이 배열 슬라이싱 하기 -> 리스트 슬라이싱과 같음
d = np.array([[1,2,3,4,5], [2,4,5,6,7], [5,7,8,9,9]])  # 3x5의 2차원 배열. 행부터 읽음.
print(d[0][1])
print(d[1:, 3:])    # 2행과, 3열 까지의 모든 배열을 가지고 온다.
```

```python
# 배열의 크기 알아내는 shape
d = np.array([2,3,4,5,6])   # 1x5의 배열
print(d.shape)
e = np.array([[1,2,3,4] , [3,4,5,6]])
print(e.shape)  # 2, 4
f = np.array([[0.1,0.2,0.3], [.2, .3, .4]])
print(f.shape)
```

```python
# 배열의 원소 유형 확인하는 dtype
print(d.dtype)
print(e.dtype)  # 이들은 함수가 아닌 필드에 가까움. 괄호가 없어서
print(f.dtype)
```

```python
# 배열의 유형 바꾸는 astype 함수
data = np.arange(1, 5) # 1~ 4 까지 이루어지는 배열 생성
print(data)
print(data.dtype)
print(data.astype('float64'))  # 기존 int 64로 된 원소들을 float64 타입으로 변경
print(data.astype('int32'))
```

```python
# 행과 열을 바꿔즈는 transpose함수
a = np.ones((2,3))
print(a)    # 2x3

b = np.transpose(a)
print(b)    # 3x2
```

```python
# (1) 배열의 사칙연산(+, -, *, /). 크기와 차원이 다른 경우. 고1때 행렬의 계산과 동일
arr1 = np.array([[2,3,4], [6,7,8]])
arr2 = np.array([[12,23,14],[36,47,58]])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
# 설문지 데이터(quest.csv) 전처리하기
import os, usecsv
import numpy as np

print(os.getcwd())
os.chdir(r'D:\python-prectice1\prectice_python\2024_06_27_thur')

quest_csv = usecsv.opencsv('quest.csv') # quest.csv파일 read
print(type(quest_csv))                  # list
print(type(quest_csv[0][0]))            # str

quest_csv = usecsv.switch(quest_csv)
print(type(quest_csv[0][0]))            # float

quest_np = np.array(quest_csv)          # numpy객체로 변경
print(quest_np)

# 넘파이 객체의 원소에서 5보다 큰 원소들의 값은 true로 변경하고 아니면 false
print(quest_np > 5)
print(quest_np[quest_np > 5])   #quest_up의 인덱싱을 활용해서 5보다 큰 수만 가져오기
# print(quest_np)

# quest_np2 = quest_np > 5
# print(quest_np[quest_np2])

quest_np[quest_np > 5] = 5   # quest_up 5보다 큰 값들은 5로 세팅 한다는 뜻.
print(quest_np)

# csv파일로 output하기
usecsv.writecsv('quest_resultcsv.csv', quest_np.astype('int32'))
```

```python
discount = 0.05 # 할인율 5퍼
cashflow = 100  # 단위는 억
```

```python
# 자본의 현재 가치 구하는 함수  -n이 커질수록, 즉 사업기간이 길면 길수록 현재가치는 떨어짐
def present_value(n):   #n은 연차.(총 사업 기간(년))
    return cashflow / ((1 + discount) ** n)     # n은 지수. **는 승.
```

```python
print(present_value(1))
print(present_value(2))
```

```python
loss = [-750]   #단위는 억원
profit = [100] * 10 #단위는 억원
print(profit)

cf = loss + profit # 리스트끼리의 '+는 insert or append개념
print(cf)

cashflow = np.array(cf)

import numpy_financial as npf   # 순 현재 가치와 내부수익률 함수를 사용하기 위한 파이썬 패키지 로드

# npf.npv(할인율, 현금흐름) -> 순현재가치
# npf.irr(현금흐름) -> 내부수익률(irr)

# #여기서 할인율은 4.5%로 가정하고 계산

npv = npf.npv(0.045, cashflow)
irr = npf.irr(cashflow)

print(npv)  # 순 현재가치는 41.3억
print(irr)  #내부 수익률 5.6%
```
